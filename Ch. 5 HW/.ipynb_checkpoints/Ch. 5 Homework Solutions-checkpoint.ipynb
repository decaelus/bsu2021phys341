#### 5.13 

$U(r) = U_\mathrm{0} \left( \frac{r}{R} + \lambda^2 \frac{R}{r}\right)$

The equilibrium position $r_\mathrm{0}$ corresponds to $\frac{dU}{dr} = 0$ and $\frac{d^2U}{dr^2} > 0$.   $\left(\frac{dU}{dr}\right)_{r_\mathrm{0}} = U_\mathrm{0} \left( \frac{1}{R} - \lambda^2 \frac{R}{r_\mathrm{0}^2}\right)\\ \Rightarrow 0 = \left( \frac{1}{R} - \lambda^2 \frac{R}{r_\mathrm{0}^2}\right) \\ \Rightarrow \frac{r_\mathrm{0}^2}{R} - \lambda^2 R = 0 \\ \Rightarrow r_\mathrm{0} = \lambda R. $

If we define $v = r/(\lambda R)$, we can re-write the potential as $U = \lambda U_\mathrm{0} \left(v + v^{-1}\right)$.

Here's a plot of the potential near the equilibrium point:

```python
#v = r/(lambda R)
v = np.linspace(0.01, 10, 1000)
U0_times_lambda = 1.0
U = U0_times_lambda*(v + 1.0/v)
asymptotic_U = U0_times_lambda*(1.0/v)
linear_U = U0_times_lambda*(v)
ax = plt.gca()
plt.plot(v, U, lw=3, color='black', label='$U(r)$')
plt.plot(v, asymptotic_U, lw=3, ls='--', label='$U \propto r^{-1}$')
plt.plot(v, linear_U, lw=3, ls='--', label='$U \propto r$')
plt.axvline(1.0, color='red', lw=3, alpha=0.5, label='equil.')

ax.set_ylim([0, 20])
ax.legend(loc='best', fontsize=18)

plt.xlabel("$r/(\lambda R)$", fontsize=18)
plt.ylabel("$U(r)$", fontsize=18)
plt.title("Potential from Problem 5.13", fontsize=18)
```

Again, we'll Taylor-expand $U(r)$ about a minimum, giving $U(r = \lambda R + x) \approx U(r = \lambda R) + \left(\frac{dU}{dr}\right)_{r = \lambda R} x + \frac{1}{2} \left(\frac{d^2 U}{dr^2}\right)_{r = \lambda R} x^2$ for $x$ small. And again, $\left(\frac{dU}{dr}\right)_{r = \lambda R} = 0$, so we just need $\left(\frac{d^2 U}{dr^2}\right)_{r = \lambda R}$: $ \left(\frac{d^2 U}{dr^2}\right)_{r = \lambda R} = +2 U_\mathrm{0} \lambda^2 \frac{R}{r_\mathrm{0}^3} \\ = +2 U_\mathrm{0} \lambda^2 \frac{R}{(\lambda R)^3} \\ = \frac{2 U_\mathrm{0}}{\lambda R^2} $

$\Rightarrow \boxed{ k = \frac{2 U_\mathrm{0}}{\lambda R^2}}$.

#### 5.21

Equation (5.28) is $\ddot{x} + 2\beta \dot{x} + \omega^2_\mathrm{0} x = 0$.<br><br>  With $x(t) = t e^{-\beta t}$, $\dot{x} = e^{-\beta t} - \beta t e^{-\beta t}$, and $\ddot{x} = -2 \beta e^{-\beta t} + \beta^2 t e^{-\beta t}$. <br><br>  Plugging into Equation (5.28), $\left( -2 \beta e^{-\beta t} + \beta^2 t e^{-\beta t} \right) + 2\beta \left( e^{-\beta t} - \beta t e^{-\beta t} \right) + \omega^2_\mathrm{0} \left( t e^{-\beta t} \right) = -\beta^2 t e^{-\beta t} + \omega^2_\mathrm{0} t e^{-\beta t}$.<br><br>  So as long as $\omega_\mathrm{0} = \beta$, Equation (5.28) is satisfied.

#### 5.22

The general solution for the position $x(t)$ of a critically damped oscillator is $x(t) = C_\mathrm{1} e^{-\beta t} + C_\mathrm{2} t e^{-\beta t}$. So solving parts (a) and (b) just involves solving for $C_\mathrm{1}$ and $C_\mathrm{2}$.

(a) $x(t = 0) = C_\mathrm{1} = 0$ and $\dot{x}(t = 0) = \left[-\beta C_\mathrm{1} e^{-\beta t} + C_\mathrm{2} e^{-\beta t} - \beta C_\mathrm{2} t e^{-\beta t} \right]_{t = 0} = C_\mathrm{2} = v_\mathrm{0}$, so $\boxed{x(t) = v_\mathrm{0} t e^{-\beta t}}$. And here's a plot of that solution:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#tprime = \beta t
t_prime = np.linspace(0, 10, 1000)

#x_prime = \beta x/v0
x_prime = t_prime*np.exp(-t_prime)

linear_x_prime = t_prime
exp_x_prime = np.exp(-t_prime)

fig = plt.figure(figsize=(6,6))

ax = plt.gca()
plt.plot(t_prime, x_prime, lw=3, color='black', label="$x(t)$")
plt.plot(t_prime, linear_x_prime, lw=3, color='red', ls='--', label="$x \propto t$")
plt.plot(t_prime, exp_x_prime, lw=3, color='blue', ls='--', label="$x \propto e^{-t}$")

ax.set_ylim([0, 0.5])

plt.legend(loc='best', fontsize=24)

plt.xlabel("$\\beta t$", fontsize=24)
plt.ylabel("$\\left(\\frac{\\beta}{v_\mathrm{0}}\\right) x$", fontsize=24)
plt.title("Position from Problem 5.22 (a)", fontsize=18)
```

(b) For this case, $x(0) = x_\mathrm{0}$ and $\dot{x}(0) = 0$, so $x_\mathrm{0} = C_\mathrm{1}$ and $0 = -\beta x_\mathrm{0} + C_\mathrm{2}$, so $C_\mathrm{2} = \beta x_\mathrm{0}$, giving $\boxed{x(t) = x_\mathrm{0} \left(1 + \beta t \right) e^{-\beta t}}$.<br><br>  After $t = 2\pi/\omega_\mathrm{0}$, $x = x_\mathrm{0} e^{-2 \pi}$.<br><br>  Here's a plot of that solution -- note the y-axis is a log scale, so the plot looks linear:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#tprime = \beta t
t_prime = np.linspace(0, 10, 1000)

#x_prime = x/x0
x_prime = (1. + t_prime)*np.exp(-t_prime)

ax = plt.gca()
plt.semilogy(t_prime, x_prime, lw=3, color='black', label="$x(t)$")
plt.axvline(2*np.pi)
plt.axhline((1. + 2*np.pi)*np.exp(-2*np.pi))

plt.text(2.*np.pi, 0.005, "$t = 2 \pi/\omega_\mathrm{0}$", rotation='vertical', fontsize=18, color='blue')
plt.text(0, 0.006, "$x = x_\\mathrm{0} \\left(1 + 2 \pi \\right) e^{-2 \pi}$", fontsize=18, color='blue')

plt.xlabel("$\\beta t$", fontsize=24)
plt.ylabel("$x/x_\mathrm{0}$", fontsize=24)
plt.title("Position from Problem 5.22 (b)", fontsize=18)
```

This equation implies $\dot{x}(t) = -\omega A \sin(\omega t - \delta) - \beta e^{-\beta t} \left[B_\mathrm{1} \cos(\omega_\mathrm{1} t) + B_\mathrm{2} \sin(\omega_\mathrm{1} t)\right] + \omega e^{-\beta t} \left[-B_\mathrm{1} \sin(\omega_\mathrm{1} t) + B_\mathrm{2} \cos(\omega_\mathrm{1} t)\right]$. 

With $x(t = 0) = x_\mathrm{0}$ and $\dot{x}(t = 0) = \dot{x_\mathrm{0}}$, I get the following relationships: $ x_\mathrm{0} = A \cos(\delta) + B_\mathrm{1}\\ \dot{x}_\mathrm{0} = \omega A \sin(\delta) - \beta B_\mathrm{1} + \omega_\mathrm{1} B_\mathrm{2}. $

And so $\boxed{B_\mathrm{1} = x_\mathrm{0} - A \cos(\delta), B_\mathrm{2} = \left(\frac{1}{\omega_\mathrm{1}}\right)\dot{x}_\mathrm{0} - \omega A \sin(\delta) + \beta B_\mathrm{1}}$.

#### 5.44

(a) The solution for $x(t) = A \cos\left(\omega t - \delta\right)$, giving a velocity $\upsilon(t) = -A \omega \sin\left(\omega t - \delta\right)$.<br><br>  The energy of the oscillator $E = \frac{1}{2} m \upsilon^2 + \frac{1}{2} m \omega^2 x^2 = \frac{1}{2} m \omega^2 A^2 \cos^2\left(\omega t - \delta\right) + \frac{1}{2} m \omega^2 A^2 \sin^2\left(\omega t - \delta\right) = \boxed{\frac{1}{2} m \omega^2 A^2}$.

(b) The energy dissipated by the damping $\Delta E$ is the work done over one cycle: <br><br>  $\Delta E = \int_P F_\mathrm{damp} \upsilon dt$, the integral over one period $P$. <br><br>  $\Delta E = \int_P 2 m \beta \upsilon\ \upsilon dt = 2 m \beta \int_P \upsilon^2 dt = 2 m \beta A^2 \omega^2 \int_P \sin^2\left(\omega t - \delta\right) dt$. <br><br>  Using Euler's identity, $\sin^2\left(\phi\right) = \left[ \frac{1}{2i}\left( e^{i \phi} - e^{-i \phi} \right) \right]^2 = -\frac{1}{4}\left[ e^{2i \phi} + e^{-2i \phi} - 2 \right] = \frac{1}{2} \left[1 - \cos\left( 2 \phi \right) \right]$, and then we can do the integral. <br><br>  With the substitution $\phi = \omega t - \delta$ (and $dt = d\phi/\omega$), the integral becomes <br><br>  $\Delta E = m \beta A^2 \omega \int_{\phi = -\delta}^{2\pi - \delta} \left[ 1 - \cos\left( 2 \phi \right) \right] d\phi = m \beta A^2 \omega \bigg[ \phi -\frac{1}{2}\sin\left( 2 \phi \right) \bigg]_{-\delta}^{2\pi - \delta} = \boxed{ 2 \pi m \beta A^2 \omega }$.

(c) $2\pi E/\Delta E = 2\pi \frac{1}{2} m \omega^2 A^2/\left( 2\pi m \omega \beta A^2 \right) = \boxed{ \omega / 2\beta = Q }$.

#### 5.53

(a) First step is to find the Fourier series representing the driving force. The force is an even function, so all the sine terms are zero, and we need only calculate the $a_\mathrm{n}$:

$a_\mathrm{n} = \frac{2}{\tau}\int_{-\tau/2}^{\tau/2} f(t) \cos(n \omega t) dt = \int_{-1}^{0} \left( 1 + t \right) \cos(n \pi t) dt - \int_{0}^{+1}\left( 1 - t \right) \cos(n \pi t) dt = \frac{2 ( 1 - \cos(n \pi) )}{(n\pi)^2}$.

As described in Section 5.8, the solution $x(t) = \sum_{n = 0}^\infty A_\mathrm{n} \cos(n \omega t - \delta_\mathrm{n})$, with $A_\mathrm{n} = \frac{f_\mathrm{n}}{\sqrt{(\omega_0^2 - n^2\omega^2)^2 + 4 \beta^2 n^2 \omega^2}}$ and $\delta_\mathrm{n} = \arctan\bigg(\frac{2 \beta n \omega}{\omega_0^2 - n^2 \omega^2}\bigg)$. 

The plot for the solution is below.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl

def fn(n):
    return 2.*(1 - np.cos(n*np.pi))/(n*np.pi)**2

def fourier_series(x, num_terms = 2, omega = np.pi):
    fs = np.ones_like(x)*0.5
    for n in range(1, num_terms + 1):
        an = fn(n)
        fs += an*np.cos(n*omega*x)
    return fs

def An(n, omega0, omega, beta):
    return fn(n)/np.sqrt((omega0**2 - (n*omega)**2)**2 + 4*(beta*n*omega)**2)

def deltan(n, omega0, omega, beta):
    ret_val = np.pi/2.
    if((omega0**2 - (n*omega)**2) != 0.):
        ret_val = np.arctan(2.*beta*n*omega/(omega0**2 - (n*omega)**2))
    
    return ret_val

def soln(t, omega0, omega, beta, num_terms = 4):
    x = np.ones_like(t)*0.5
    
    for n in range(1, num_terms + 1):        
        x += An(n, omega0, omega, beta)*np.cos(omega*t - deltan(n, omega0, omega, beta))
        
    return x

def driving_force(t, tau, fmax = 1.):
    t_mod = (t % tau) - tau
   
    ret_val = np.zeros_like(t_mod)
    ret_val[(t_mod > -1.0)] = fmax*t_mod[(t_mod > -1.0)] + 1.0
    ret_val[(t_mod < -1.0)] = -fmax*(t_mod[(t_mod < -1.0)] + 2.) + 1.0

    return ret_val

tau0 = 2.
omega0 = 2.*np.pi/tau0
beta = 0.1
#driving period
tau = 2.
omega = 2.*np.pi/tau

fig = pl.figure(figsize=(8,8))
ax = fig.add_subplot(111)

t = np.linspace(-1, 3, 100)
ax.plot(t, driving_force(t, tau), label='driving force', lw=3)
print("solution -- ")
ax.plot(t, fourier_series(t, num_terms=4), label='driving force (Fourier series)', 
        lw=6, ls='--', alpha=0.5)
x = soln(t, omega0, omega, beta)
ax.plot(t, x, lw = 3, ls='-.', label='oscillator')
ax.set_xlabel('time', fontsize=18)
ax.legend(loc='best', fontsize=18)
ax.tick_params(axis='both', labelsize=16)
```
