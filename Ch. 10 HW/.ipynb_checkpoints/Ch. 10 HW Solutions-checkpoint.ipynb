(a) We can find $z$-component of the center of mass for our bubble $z_{\rm bubble}$ by using subtracting the center of mass for a hemisphere of radius $a$ weighted by its mass, $M_{\rm a} z_{\rm a}$ from that of a hemisphere of radius $b$, $M_{\rm b} z_{\rm b}$ and then dividing through by the mass of the bubble, $M_{\rm bubble} = M_{\rm b} - M_{\rm a}$, i.e. <br>

$z_{\rm bubble} = \dfrac{M_{\rm b} z_{\rm b} - M_{\rm a} z_{\rm a}}{M_{\rm b} - M_{\rm a}} = \frac{3}{8}\left(\dfrac{b^4 - a^4}{b^3 - a^3}\right)$

(b) $ z_{\rm bubble} = \frac{3}{8} b$, i.e. just a hemisphere with radius $b$.

(c) If we define $b = a + \delta a$ and take the limit as $\delta a \rightarrow 0$, we'll get the result desired.<br>

$b^4 = \left( a + \delta a\right)^4 \rightarrow a^4 \left( 1 + 4 \delta a/a \right)$<br>
$b^{-3} = \left( a + \delta a\right)^{-3} \rightarrow a^{-3}\left(1 - 3 \delta a/a \right)$

$z_{\rm bubble} \approx \frac{3}{8} a \left( 1 + 4 \delta a\ a \right) \left(1 - 3 \delta a/a \right) \rightarrow \boxed{\frac{3}{8} a}$.

(a) Using the definition $I_{\rm xx} = \int_{\rm V} \rho \left(r^2 - x^2 \right) dV = \int_{\rm V} \rho \left( y^2 + z^2 \right) dV$. If we shift $y \rightarrow y + \eta$ and $z \rightarrow z + \zeta$, we get $I_{\rm xx} = \int_{\rm V} \rho \left( \left( y + \eta \right)^2 + \left( z + \zeta \right)^2 \right) dV = \int_{\rm V} \rho \left( \left( y^2 + \eta^2 + 2 y \eta \right) + \left( z^2 + \zeta^2 + 2 z \zeta \right) \right) dV$. <br>  The cross terms $2 y \eta$ and $2 z \zeta$ integrate to zero since $y$ and $z$ are measured relative to CM. The $\left( y^2 + z^2 \right)$ integral is just $I^{\rm CM}_{\rm xx}$, and the $\left( \eta^2 + \zeta^2 \right)$ integral is just $M \left( \eta^2 + \zeta^2 \right) $ since $\eta$ and $\zeta$ are constants. <br><br>  Next, $I_{\rm yz} = -\int_{\rm V} \rho \left( y + \eta \right) \left( z + \zeta \right) dV = -\int_{\rm V} \rho \left( yz + y\zeta + z\eta + \eta \zeta \right) dV = I^{\rm CM}_{\rm yz} - M \eta \zeta$ since all the cross terms with $y$ and $z$ integrate to zero.

(b) For a cube $I^{\rm CM}_{xx} = \frac{1}{6} M a^2$, and Example 10.2 has $\xi = \eta = \zeta = -\frac{1}{2} a$. From (a), $I_{\rm xx} = I^{\rm CM}_{xx} + M \left( \eta^2 + \zeta^2 \right) = \frac{1}{6} M a^2 + \frac{2}{4} M a^2 = \frac{2}{3} M a^2$. Employ a similar procedure to calculate the other moment components.

#### 10.25

(a) The off-diagonal elements ($I_{\rm i \ne j}$) are zero by symmetry.<br><br>  $I_{\rm xx} = \int \rho \left( y^2 + z^2 \right) dV = \int_{\rm x = -a}^{+a} dx \int_{\rm y = -b}^{+b} dy \int_{\rm z = -c}^{+c} dz\ \rho \left( y^2 + z^2 \right) = \rho \left( 2a \right) \left[ \frac{2}{3}b^3 \left(2c\right) + \frac{2}{3}c^3 \left(2b\right) \right] = \frac{1}{3} \rho 2a\ 2b\ 2c \left[ b^2 + c^2 \right] = \boxed{ \frac{1}{3} M \left[ b^2 + c^2 \right] }$.<br><br>  For the other diagonal elements, just change out $b$ and $c$ in the last term.

(b) Using the parallel axis theorem, we can calculate $I_{\rm ij}$ for the case that the origin is shifted by $\vec{\Delta} = (-a, -b, -c)$, i.e. to one of the cuboid's corners.<br><br>  $I_{\rm xx} = I_{\rm xx}^{\rm CM} + M \left( b^2 + c^2 \right) = \boxed{ \frac{4}{3} M \left( b^2 + c^2 \right) }$, and $I_{\rm xy} = I_{\rm xy}^{\rm CM} - M \left( -a \right) \left( -b \right) = \left( 0 \right) - M a b = \boxed{ -M a b }$. <br><br>  Symmetry allows us to easily extrapolate and calculate the whole tensor: <br><br> $ {\bf I} = M \left| \begin{array}{ccc} \frac{4}{3} \left( b^2 + c^2 \right) & -ab & -ac \\ -ab & \frac{4}{3} \left( a^2 + c^2 \right) & -bc \\ -ac & -bc & \frac{4}{3} \left( a^2 + b^2 \right) \end{array} \right|.$

(c) $\vec{\omega} = \left( \begin{array}{c} \omega \\ 0 \\ 0 \end{array} \right)$, and $\vec{L} = {\bf I} \cdot \vec{\omega} = \boxed{ M \left[ \frac{4}{3} \left( b^2 + c^2 \right) - ab - ac \right] \omega }$.
