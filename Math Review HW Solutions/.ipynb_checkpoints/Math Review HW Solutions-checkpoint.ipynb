#### p. 169, 73
The differential equation is $\dfrac{d^2 y}{dt^2} + \omega^2 y = 0$.<br>

Taking $y_{\rm 1} = A \sin \omega t$ and $y_{\rm 2} = B \cos \omega t$, so that $y = y_{\rm 1} + y_{\rm 2}$.<br>
$\dfrac{dy_{\rm 1}}{dt} = A \omega \cos \omega t$ and $ \dfrac{d^2y_{\rm 1}}{dt^2} = -A \omega^2 \sin \omega t $.<br>
Similarly, for $y_{\rm 2}$: $\dfrac{dy_{\rm 2}}{dt} = -B \omega \sin \omega t$ and $ \dfrac{d^2y_{\rm 1}}{dt^2} = -B \omega^2 \cos \omega t $.<br><br>

Plugging back into the differential equation: $\bigg( \dfrac{d^2y_{\rm 1}}{dt^2} + \dfrac{d^2y_{\rm 1}}{dt^2} \bigg) + \omega^2 \bigg( y_{\rm 1} + y_{\rm 2} \bigg) = \bigg( -A \omega^2 \sin \omega t - B \omega^2 \cos \omega t \bigg) + \omega^2 \bigg( A \sin \omega t + B \cos \omega t \bigg) = 0$, so the equation is satisfied.

#### p. 298, 54
A constant, negative acceleration means $\dfrac{d^2x}{dt^2} = -a.$ To work out the distance traveled in time $t$ under such an acceleration, we need to integrate twice.<br><br>

$$
\dfrac{dx}{dt}(t) = \int_{t^\prime = 0}^{t} \dfrac{d^2x}{dt^2}\ dt = -\int_{t^\prime = 0}^{t} a\ dt = -a\ t + v(0).
$$ <br>

$$
x(t) = \int_{t^\prime = 0}^{t} \dfrac{dx}{dt}\ dt = \int_{t^\prime = 0}^{t} \big( -a\ t + v(0) \big) dt = -\frac{1}{2} a t^2 + v(0)\ t + x(0).
$$

We can take $x(0) = 0$ for simplicity, giving $x(t) = v(0) t - \frac{1}{2} a t^2$.

At $t = 0$, the carts are separated by distance $s$, and after time $t$, the second cart will be at a distance $v_{1} t + s$. To just avoid collision, the distance traveled by the first cart has to be equal to that, i.e.:

$$
x(t) = v(0) t - \frac{1}{2} a t^2 = s + v_1 t.
$$

The other condition to avoid a collision is that the velocity of the first cart $v(t) = v(0) - a t$ cannot exceed the velocity of the second after time $t$, i.e. $v(0) - a t = v_1.$ This equation lets us solve for $t$:
$$
t = \dfrac{v(0)- v_1}{a}.
$$

We can plug this into our first equation for $x(t)$ and solve for $v(0)$:
$$
v(0) \left( \dfrac{v(0) - v_1}{a} \right) - \frac{1}{2} a \left( \dfrac{v(0) - v_1}{a} \right)^2 = s + v_1 \left( \dfrac{v(0) - v_1}{a} \right).
$$
giving

$\frac{1}{2} v^2(0) - v_{1} v(0) + \left( \frac{1}{2} v_1^2 - a s \right) = 0$.

Using the quadratic equation gives $v(0) = v_1 \pm \sqrt{2 a s}.$ Since $v(0) > v_1$ by assumption, we will take the positive root, and
$$
\boxed{ v(0) = v_1 + \sqrt{2 a s} }.
$$

#### p. 308, 76
The area of the circle is $A = 4 \int_{x = 0}^{r} \sqrt{r^2 - x^2}\ dx.$ Easier to integrate if we use the substitution $x = r \sin \theta$. In this case, the circle's radius $r$ is a constant, and so $dx = r \cos \theta\ d\theta$. For the lower limit, $x = 0$, $\theta = 0$. For the upper limit, $x = r$, $\theta = \pi/2$. Thus,

$$
A = 4\ \int_{\theta = 0}^{\pi/2} \sqrt{r^2 - r^2 \sin^2 \theta}\ r \cos \theta\ d\theta\\
  = 4\ \int_{\theta = 0}^{\pi/2} r\sqrt{1 - \sin^2 \theta}\ r \cos \theta\ d\theta \\
  = 4\ \int_{\theta = 0}^{\pi/2} r\sqrt{\cos^2 \theta}\ r \cos \theta\ d\theta \\
  = 4 r^2 \int_{\theta = 0}^{\pi/2} \cos^2\theta\ d\theta.
$$

There are lots of ways to calculate this integral, but perhaps the most straightforward is to use a trig identity: $\cos^2 \theta = \dfrac{1 + \cos\left( 2 \theta \right)}{2}.$

$$
A = 4\ r^2 \int_{\theta = 0}^{\pi/2} \dfrac{1 + \cos\left( 2 \theta \right)}{2} d\theta\\
  = 2\ r^2 \bigg[ \theta + \frac{1}{2} \sin\left( 2 \theta \right) \bigg]_{0}^{\pi/2}\\
  = \pi r^2.
$$

#### p. 42, 19
Remember the Taylor series expansion around $x = a$ for a well-behaved function $f(x)$ is 
$$
f(x) = f(a) + f^\prime(a)\ \left( x - a \right) + \dfrac{f^{\prime\prime}(a)}{2!}\ \left( x - a \right)^2 + \dfrac{f^{\prime\prime\prime}(a)}{3!}\ \left( x - a \right)^3 + \ldots
$$

For $f(x) = \sin x$ expanded about $a = \pi$, we get
$$
\sin x \approx \sin \pi + \left( \cos \pi \right) (x - a) - \dfrac{\sin \pi}{2} (x - a)^2 - \dfrac{\cos \pi}{3!} (x - a)^3 \\
       = \boxed{ -(x - a) + \dfrac{(x - a)^3}{6} }.
$$
I compare the two solutions below.

```python
def Taylor_expansion(x):
    return -(x - np.pi) + (x - np.pi)**3/6

x = np.linspace(np.pi/2, 3.*np.pi/2, 100)

fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(111)

ax1.plot(x/np.pi, np.sin(x), lw=3, color="blue", label="$\sin\ x$")
ax1.plot(x/np.pi, Taylor_expansion(x), lw=6, ls='--', color="blue", label="Taylor")

ax2 = ax1.twinx()
ax2.plot(x/np.pi, (np.sin(x) - Taylor_expansion(x))/np.sin(x)*100, lw=3, ls='-.', color="red", label="% error")

ax1.set_xlabel("$x/\pi$", fontsize=24)
ax1.set_ylabel("functions", fontsize=24, color="blue")
ax1.tick_params('both', labelsize=18)
ax1.tick_params('y', labelcolor="blue")

ax2.set_ylabel("% error", fontsize=24, color="red")
ax2.tick_params('both', labelsize=18, labelcolor="red")
```

#### p. 295, 17
Perform line integral $\int \vec{F} \cdot d\vec{r}$ over this circle:

We are given $x^2 + y^2 = 2$, which represents a circle of constant radius $r = \sqrt{2}$. Thus, we can re-cast $x$ and $y$ using polar coordinates: $x = \sqrt{2} \cos \theta$ and $y = \sqrt{2} \sin \theta$. 

Then at point $(1, 1)$, we have $\theta = \pi/4$ (45$^\circ$), and at $(1, -1)$, we have $\theta = -\pi/4$.

We need also to convert our arc length vector $d\vec{r}$ into polar coordinates: $d\vec{r} = dx\ \hat{i} + dy\ \hat{j} = \sqrt{2} \left( -\sin \theta\ \hat{i} + \cos \theta\ \hat{j} \right) d\theta$.

Then converting the vector field, $\vec{F} = \left( 2\sqrt{2} \cos \theta - 3 \sqrt{2} \sin \theta \right)\ \hat{i} - \left( 3\sqrt{2} \cos \theta - 2\sqrt{2} \sin \theta \right)\ \hat{j}$.

Finally we calculate the integral: 
$$
\int_{\theta = \pi/4}^{-\pi/4} \bigg[ \left( 2\sqrt{2} \cos \theta - 3 \sqrt{2} \sin \theta \right)\ \hat{i} - \left( 3\sqrt{2} \cos \theta - 2\sqrt{2} \sin \theta \right)\ \hat{j} \bigg] \cdot \bigg[ \sqrt{2} \left( -\sin \theta\ \hat{i} + \cos \theta\ \hat{j} \right) \bigg] d\theta\\
= 2 \int_{\theta = \pi/4}^{-\pi/4} \bigg(-2 \cos \theta\ \sin \theta + 3 \sin^2\theta - 3 \cos^2 \theta + 2 \sin\theta\ \cos \theta \bigg) d\theta\\
= 6 \int_{\theta = \pi/4}^{-\pi/4} \bigg( \sin^2 \theta - \cos^2 \theta \bigg) d\theta
$$

We can substitute the relationship $\sin^2 \theta = 1 - \cos^2\theta$, which gives us
$$
I = 6 \int_{\theta = \pi/4}^{-\pi/4} \bigg( 1 - 2 \cos^2 \theta \bigg) d\theta.
$$

And then we can use the identity from above $\cos^2 \theta = \dfrac{1 + \cos\left(2 \theta\right)}{2}$, giving
$$
= -6 \int_{\theta = \pi/4}^{-\pi/4} \cos\left( 2 \theta \right)\ d\theta\\
= -6 \bigg[ \frac{1}{2} \sin \left( 2 \theta \right) \bigg]_{\pi/4}^{-\pi/4}\\
=-3 \bigg[\sin \left(-\pi/2\right) - \sin \left(\pi/2\right) \bigg] = \boxed{ 6 }.
$$
