A mass density $\rho$ means every tiny bit of fluid volume $\Delta V$ contains mass $\Delta M \approx \rho \Delta V$. So with a velocity $\upsilon$, through what volume of fluid does the projectile pass from time $t_0$ to time $t_0 + \Delta t$? The figure below illustrates the geometry.

```python
Image(filename='HW2_Fig2p4a2.jpg', width=500)
```

With $\Delta t$ is small enough that $\upsilon \approx {\rm const.}$ during that time, the distance traveled in $\Delta t$ is $\Delta L \approx \upsilon \Delta t$, and so the projectile sweeps through a volume $\Delta V \approx A \Delta L = A \upsilon \Delta t$.

Thus, the mass of fluid swept up is $\Delta M \approx \rho A \upsilon \Delta t \Rightarrow \boxed{ \dfrac{\Delta M}{\Delta t} = \rho A \upsilon}$.

### An Aside

How did I know to assume $\upsilon \approx {\rm const.}$ during $\Delta t$? We know the projectile's velocity is dropping due to drag at every instant, so why can I assume it's constant at all?

Remember we're assuming $\Delta t$ is small. What we really mean by that is that the expressions we develop will be more accurate the smaller $\Delta t$ is. In reality, for any finite $\Delta t$, $\upsilon$ changes over that time interval. In the end, though, we'll end up taking the limit as $\Delta t \rightarrow 0$ anyway, so our approximation will be arbitrarily accurate.

But let's say we didn't assume $\upsilon$ was constant over $\Delta t$ but that it changed just a little bit. In that case, the velocity at $t_0 + \Delta t$ is approximately $\upsilon(t_0 + \Delta t) \approx \upsilon(t_0) + \dot{\upsilon}_{t = t_0} \Delta t$. In other words, we'll Taylor-expand $\upsilon$ about $t_0$.

Then if I go back to my expression for $\Delta M$, $\Delta M \approx \rho A \left( \upsilon + \dot{\upsilon} \Delta t \right) \Delta t = \rho A \upsilon \Delta t + \rho A \dot{\upsilon} \left(\Delta t\right)^2$. The second term will be much smaller than the first, provided that $\Delta t \ll \upsilon/\dot{\upsilon}$.

(b) If the projectile has to accelerate all the fluid it encounters from rest to a velocity roughly equal to its own at some instant in time, then the projectile is constantly shedding momentum into the fluid, and conservation of momentum requires the projectile decelerate.

So conservation of momentum for the whole system requires $P_{\rm before} = P_{\rm after}$, where we're considering the momentum before and after the projectile has passed through a small volume of fluid. 

Initially the fluid is at rest, and $P_{\rm before} = M_{\rm proj} \upsilon(t)$. Afterward, the projectile AND fluid are in motion, so $P_{\rm after} \approx M_{\rm proj} \upsilon(t + \Delta t) + \Delta M \upsilon(t + \Delta t)$.

With $\Delta t$ small and $\Delta M \propto \Delta t$, we can assume $\Delta M$ has a velocity $\upsilon(t)$ and ignore the second-order term.

$M_{\rm proj} \upsilon(t) \approx M_{\rm proj} \left(\upsilon + \dot{\upsilon} \Delta t \right) + \Delta M \upsilon$ <br>
$\Rightarrow 0 \approx M_{\rm proj} \dot{\upsilon} \Delta t + \Delta M \upsilon = M_{\rm proj} \dot{\upsilon} \Delta t + \rho A \upsilon^2 \Delta t$.

Thus, $\boxed{ M_{\rm proj} \dot{\upsilon} = -\rho A \upsilon^2 }$.

Other factors of order unity are required to account for imperfect transfer of momentum.

(c) Eqn (2.3) says $f_{\rm quad} = c \upsilon^2 = \kappa \rho A \upsilon^2$.

$A = \pi r^2 = \pi \left( \frac{1}{2} D \right)^2 = \frac{\pi}{4} D^2$.

So if $c = \gamma D^2$ and $\gamma = \frac{\pi}{4} \kappa \rho$, then our result for (b) agrees with Eqn (2.3).

$\gamma = \frac{\pi}{4} \left( 1/4 \right) \left( 1.29\ {\rm kg/m^3} \right) = 0.25\ {\rm kg/m^3}$.

## 2.10

(a) The forces on the ball bearing are $m \ddot{y} = ({\rm weight}) + ({\rm buoyancy}) + ({\rm linear\ drag})$. How to calculate buoyancy?

To move downward, the ball bearing must displace a mass of glycerin $V_{\rm gl}$ equal to its own volume, $V_{\rm bb}$, i.e. $V_{\rm bb} = V_{\rm gl}$. That essentially requires that the glycerin is lifted against gravity, which requires the BB to apply a force equal to the weight of the displaced glycerin. Of course, the glycerin applies an equal and opposite force, so the buoyant force is just the weight of the displaced fluid.

The weight of the glycerin $W_{\rm gl} = \rho_{\rm gl} V_{\rm gl} g = \rho_{\rm gl} V_{\rm bb} g$.

Thus, $m \ddot{y} = -m g + \rho_{\rm gl} V_{\rm bb} g + \beta D \dot{y}$, and $m = \rho_{\rm bb} V_{\rm bb}$ and $V_{\rm bb} = \frac{\pi}{6} D^3$. 

Plugging these in and solving the equation for $\ddot{y}$ gives $\ddot{y} = -\left( \dfrac{\rho_{\rm bb} - \rho_{\rm gl}}{\rho_{\rm bb}} \right) g + \dfrac{\beta D}{m_{\rm bb}} \dot{y}$. So it's the same equation as in class, just with a smaller gravity: $\left( \dfrac{\rho_{\rm bb} - \rho_{\rm gl}}{\rho_{\rm bb}} \right) = \left( \dfrac{7.8\ {\rm g/cm^3} - 1.3\ {\rm g/cm^3}}{7.8\ {\rm g/cm^3}} \right) = 0.83$, or $g^\prime = 0.83\ g$.

From problem (2.2), $\upsilon_{\rm ter} = \dfrac{m g^\prime}{b} = \dfrac{m g^\prime}{3\pi \eta D} \Rightarrow \upsilon_{\rm ter} = \dfrac{\rho_{\rm bb} D^2 g^\prime}{18 \eta} = \dfrac{\left(7.8\ {\rm g/cm^3}\right) \left( 2\ {\rm mm} \right)^2 \left( 0.83 \right) \left(9.8\ {\rm m/s^2} \right)}{18\ \left( 12\ {\rm N\ s/m^2} \right)} = \boxed{ 0.1\ {\rm cm/s} }$.

And $\tau = \upsilon_{\rm ter}/g^\prime = \boxed{ 120\ {\rm \mu s} }$.

From p. 52, we see the BB should reach 95\% of $\upsilon_{\rm ter}$ after $3\tau = \boxed{ 360\ {\rm \mu s} }$.

(b) $\dfrac{f_{\rm quad}}{f_{\rm lin}} = \dfrac{\rho_{\rm gl} D \upsilon}{48 \eta} \sim 10^{-6}$, so the approximation is good to a few parts in a millon.

## 2.28

$m \ddot{x} = -c \upsilon^{3/2} \rightarrow m \dot{\upsilon} = -c \upsilon^{3/2}$.

$\dot{\upsilon} = \dfrac{d\upsilon}{dt} = \dfrac{d\upsilon}{dx} \dfrac{dx}{dt} = \upsilon \dfrac{d\upsilon}{dx} \Rightarrow m \upsilon \dfrac{d\upsilon}{dx} = -c \upsilon^{3/2} \Rightarrow m \dfrac{d\upsilon}{dx} = -c \upsilon^{1/2} $.

$-\dfrac{m}{c} \dfrac{d\upsilon}{\upsilon^{1/2}} = dx \rightarrow -\int_{\upsilon = \upsilon_0}^{0} \dfrac{m}{c} \dfrac{d\upsilon}{\upsilon^{1/2}} = \int_{x = 0}^{x} dx$

$\Rightarrow x = \dfrac{2m}{c} \upsilon_0^{1/2}$.

(a) $z = e^{i \theta}$

$z^2 = \left( e^{i \theta} \right)^2 = e^{2 i \theta} = \cos\left( 2 \theta \right) + i \sin\left( 2 \theta \right)$ 

and

$z^2 = \big[ \cos\left( \theta \right) + i \sin\left( \theta \right) \big]^2 = \cos^2 \theta - \sin^2 \theta + 2i \cos\theta \sin\theta$. 

Equating the real and imaginary parts:
$\cos\left( 2 \theta \right) = \cos^2 \theta - \sin^2 \theta$ and $\sin\left( 2 \theta \right) = 2 \cos \theta \sin \theta$

(b) $z^3 = e^{3i\theta} = \cos\left( 3 \theta \right) + i \sin\left( 3 \theta \right)$.

and 

$\big( \cos \theta + i \sin \theta \big)^3 = \cos^3 \theta - \cos\theta \sin^2 \theta - 2 \cos \theta \sin^2 \theta + i \big( \cos^2 \theta \sin \theta - \sin^3 \theta + 2 \cos^2 \theta \sin\theta \big) $.

Again, equating real and imaginary parts:
$\cos \left( 3 \theta \right) = \cos \theta \left( \cos^2 \theta - 3 \sin^2 \theta \right)$ and 
$\sin \left( 3 \theta \right) = \sin \theta \left( 3 \cos^2 \theta - \sin^2 \theta \right)$.
