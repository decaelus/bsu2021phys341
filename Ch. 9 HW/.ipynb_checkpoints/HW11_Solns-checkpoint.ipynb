#### 9.3

(a) $\vec{F}_{\rm tide} = -G M_{\rm moon} m \left( \dfrac{\hat{d}}{d^2} - \dfrac{\hat{d}_0}{d_0^2}\right)$. At point P, $\hat{d} = \hat{d}_0 \Rightarrow \vec{F}_{\rm tide} = -G M_{\rm moon} m \left( \dfrac{1}{d^2} - \dfrac{1}{d_0^2}\right) \hat{d}_0$.<br><br>  And $d^{-2} = \left( d_0 - R_{\rm Earth} \right)^{-2} = d_0^{-2} \left[ 1 - \left( \dfrac{R_{\rm Earth}}{d_0} \right) \right]^{-2} \approx d_0^{-2} \left[ 1 + 2 \left( \dfrac{R_{\rm Earth}}{d_0} \right) \right]$ since $R_{\rm Earth} \ll d_0$.<br><br>  So $\vec{F}_{\rm tide} \approx -G M_{\rm moon} m d_0^{-2} \left[ 1 + 2 \left( \dfrac{R_{\rm Earth}}{d_0} \right) - d_0^{-2} \right] \hat{d}_0 = \boxed{ -2\dfrac{G M_{\rm moon} m}{d_0^2} \left( \dfrac{R_{\rm Earth}}{d_0} \right) \hat{d}_0}$.

The Earth's surface gravity points along the opposite direction and compared to the tidal force:
$|F_{\rm tide}|/mg = 2 \dfrac{G M_{\rm Moon} m}{d^2_{0}}\left(\dfrac{R_{\rm Earth}}{d_{0}} \right)/\left(m \dfrac{G M_{\rm Earth}}{R_{\rm Earth}^2}\right) = 2\left( \dfrac{M_{\rm Moon}}{M_{\rm Earth}}\right) \left( \dfrac{R_{\rm Earth}}{d_{0}} \right)^3 \sim 2 \left( 1/81 \right) \left( 1/60 \right)^3 \sim 10^{-7}$ -- very small.

(b) The only difference between point R and P is that at R $d^{-2} = \left( d_0 + R_{\rm Earth} \right)^{-2} \Rightarrow \boxed{ \vec{F}_{\rm tide} \approx +2\dfrac{G M_{\rm moon} m}{d_0^2} \left( \dfrac{R_{\rm Earth}}{d_0} \right) \hat{d}_0}$

#### 9.9

The Coriolis acceleration is $\vec{a}_{\rm cor} = 2 \dot{\vec{r}} \times \vec{\Omega}$, with $\vec{\Omega} = \Omega \left( \hat{y} \sin \theta + \hat{z} \cos \theta \right)$ and $\dot{\vec{r}} = \hat{y} v$. So $\vec{a}_{\rm cor} = 2 \hat{y} v \times \Omega \left( \hat{y} \sin \theta + \hat{z} \cos \theta \right) = \boxed{ 2 v \Omega \cos \theta \hat{x} }$.<br><br>  For $\theta = 40^\circ$ and $v = 1000$ m/s, $a_{\rm cor} = 2\ \left({\rm 1,000\ m/s}\right) \left( \dfrac{2 \pi}{\rm 86,400\ s} \right) \cos 40^\circ = \boxed{ 0.1\ {\rm m/s}^2 }= 1\%$ of gravity.

#### 9.27

In the rotating frame, the falling projectile experiences a Coriolis acceleration to the east because $\dot{\vec{r}} \times \vec{\Omega}$ points along $\hat{x}$. <br><br>  In the inertial frame, both the observer on the Earth's surface and the projectile (before its dropped) are rotating with angular velocity $\Omega$, but the projectile has a linear velocity that is larger by $\Delta v = \Omega h$, where $h$ is the height from which the projectile is dropped. <br><br>  As a consequence, the projectile has a higher angular momentum about the Earth's center. As it falls, conservation of angular momentum increases its tangential velocity into the east, so it lands farther east than the observer at the Earth's surface is.

#### 9.29

(a) $\vec{v}(t = 0) = v_0 \hat{z}$. Taking the dynamical equations given in Chapter 9 (p. 353),  $$\begin{eqnarray} \ddot{x} & = & 2 \Omega  \left( \dot{y} \cos \theta - \dot{z} \sin \theta \right),\\ \ddot{y} & = & -2 \Omega \dot{x} \cos \theta,\\ \ddot{z} & = & -g + 2 \Omega \dot{x} sin \theta. \end{eqnarray}$$  At the zeroth order, all the terms involving $\Omega$ are negligible, giving  $$\begin{eqnarray} \ddot{x}_0 & \approx & 0 \Rightarrow x_0(t) = 0,\\ \ddot{y}_0 & \approx & 0 \Rightarrow y_0(t) = 0,\\ \ddot{z}_0 & \approx & -g \Rightarrow z_0(t) = v_0 t - \frac{1}{2} g t^2. \end{eqnarray}$$  Plugging these solutions back into the original equations gives us the dynamics for the next order solution:  $$\begin{eqnarray} \ddot{x}_1 & \approx & -2 \Omega \dot{z}_0 \sin \theta = -2 \Omega \sin \theta \left( v_0 - g t \right) \Rightarrow x_1(t) = -\Omega v_0 \sin \theta t^2 + \frac{1}{3} \Omega g \sin \theta t^3,\\ \ddot{y}_1 & \approx & 0 \Rightarrow y_1(t) = 0,\\ \ddot{z}_1 & \approx & -g \Rightarrow z_1(t) = v_0 t - \frac{1}{2} g t^2. \end{eqnarray}$$  Solving for the time $t$ at which the ball hits the ground ($z_1 = 0$) gives $t = 2 v_0/g$.<br><br>  At that time, $x(t) = -\Omega v_0 \sin \theta \left( 2 v_0/g \right)^2 + \frac{1}{3} \Omega g \sin \theta \left( 2 v_0/g \right)^3 = \boxed{ -\left( 4 \Omega \sin \theta\ v_0^3\right)/ \left( 3 g^2 \right) }$.

(b) At the equator, $\theta = \pi/2$, and we're told $v_0 = 40$ m/s (90 mph!!). Then $x = -\left[ 4 \left( \frac{2 \pi}{\rm 86,400\ s} \right) \left( {\rm 40\ m/s } \right)^3 \right]/\left[3 \left( {\rm 9.8\ m/s^2} \right)^2 \right] = \boxed{ 6.5\ {\rm cm} }$. Small but measurable.

(c) The observer just to the north will also see the ball land 6.5 cm west of its original position. S/he sees the same Coriolis acceleration, even though her/his $\theta$ is smaller than the person to the south, because the $xyz$-velocity components of the ball are different than for the person to the south. As a consequence, the $xyz$-accelerations turn out to be the same.<br><br>  The thrown ball lands to the west while the dropped ball lands to the east because the dropped ball has a higher initial angular momentum than the thrown ball. The thrown ball's tangential (eastward) velocity starts out large, drops during its flight, then returns to its original value just as it lands. In the meantime, the Earth's surface has rotated into the east at a constant velocity, and so the thrown ball lags behind and lands to the west.
