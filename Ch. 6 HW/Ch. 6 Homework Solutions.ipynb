#### 6.22

For this problem, we need to find the relationship $y = y(x)$ that maximizes the area integral $A = \int_{0}^{x_1} y(x)\ dx$, subject to the constraint that the total arc length $S = \int ds = l = {\rm const.}$

 This constraint makes it difficult to directly calculate the area integral as an integral over $x$ because the total length will depend on both $x$ and $y$, so the endpoint $x_1$ would actually depend on our choice of $y(x)$. 

Taylor suggests side-stepping this difficulty by changing the integration variable from $x$ to $s$, i.e., the total arc length. In other words, we want to consider this integral $A = \int_{s = 0}{l} f(s)\ ds$, where $f(s)$ is some expression for the area integrand that depends on $s$ instead of $x$.

The first thing to do is to convert $dx$ into an expression in terms of $ds$:
$$
ds^2 = dx^2 + dy^2 \Rightarrow dx = \sqrt{ds^2 - dy^2} = \sqrt{1 - y^{\prime 2}} ds,
$$
where $y^{\prime} = \frac{dy}{ds}$

Then we plug this into the area integral:
$$
A = \int_{s = 0}^{l} y(s) \sqrt{1 - y^{\prime 2}} ds,
$$
and we're there! Now we can use our Euler-Lagrange equation to solve for $y = y(s)$. Taylor suggests using a slightly different form of the equation:
$$
f - y^\prime \dfrac{\partial f}{\partial y^\prime} = {\rm const.},
$$
where $f = y \sqrt{1 - y^{\prime 2}}$.

$$
y \sqrt{1 - y^{\prime 2}} + y^{\prime} \dfrac{y y^{\prime}}{\sqrt{1 - y^{\prime 2}}} = {\rm const.}
$$

Taylor indicates that the relationship $y = y(s)$ that will satisfy this equation is a semi-circle, so let's look at how to write the $y$-coordinate of a circle in terms of arc length $s$. 

For a circle, we have $x = r \cos \theta$ and $y = r \sin \theta$, where $r$ is the circle's radius which is a constant. Therefore, $dy = r \cos \theta\ d\theta$. We also have $ds = r d\theta$, so 
$$
ds = r d\theta = \dfrac{dy}{\cos \theta} \Rightarrow \dfrac{dy}{ds} = \cos \theta.
$$

We can plug these expressions for $y$ and $y^\prime$ into the Euler-Lagrange equation:
$$
y \sqrt{1 - y^{\prime 2}} + y^{\prime} \dfrac{y y^{\prime}}{\sqrt{1 - y^{\prime 2}}} = r \sin \theta \sqrt{1 - \cos^2\theta} + \cos \theta \dfrac{r\sin \theta \cos \theta}{\sqrt{1 - \cos^2\theta}} = r \sin^2\theta + r \cos^2 \theta = r = {\rm const.}
$$

(a) The plane's ground speed $\vec{\upsilon}_\mathrm{ground}$ is related to its air speed $\vec{\upsilon}_0$ and the speed of the wind $\vec{\upsilon}_\mathrm{wind}$ as $\vec{\upsilon}_\mathrm{ground} = \vec{\upsilon}_0 + \vec{\upsilon}_\mathrm{wind} = \upsilon_0\left( \cos \phi \hat{x} + \sin \phi \hat{y} \right) + V y \hat{x} = \upsilon_0 \cos \phi \left( \hat{x} + \tan \phi \hat{y} \right) + V y \hat{x} = \boxed{ \upsilon_0 \cos \phi \left( \hat{x} + y^\prime \hat{y} \right) + V y \hat{x} }$.

(b) We can write a little time $dt = ds/\upsilon_{\rm ground}$, where $ds = \sqrt{dx^2 + dy^2} = \sqrt{1 + y^{\prime\ 2} } dx$ and $\upsilon_{\rm ground} = \sqrt{ \left( \upsilon_0 \cos \phi + V y \right)^2 + \upsilon_0^2 \sin^2 \phi }$.

So 
$$
dt = \frac{\sqrt{ 1 + y^{\prime\ 2} } }{\sqrt{ \left( \upsilon_0 \cos \phi + Vy \right)^2 + \upsilon_0^2 \sin^2 \phi}} dx.
$$

We can simplify the bottom expression a bit: $\left( \upsilon_0 \cos \phi + Vy \right)^2 + \upsilon_0^2 \sin^2 \phi = \upsilon_0^2 + \left( Vy \right)^2 + 2 \upsilon_0 \cos \phi\ V y.$

We're told we can assume $\phi$ and $y^\prime$ are both small, so $\cos \phi \approx 1$ and $\sqrt{ 1 + y^{\prime\ 2} } \approx \left( 1 + \frac{1}{2} y^{\prime\ 2} \right)$. 

We can approximate $dt$ as 
$$
dt \approx \frac{\left( 1 + \frac{1}{2} y^{\prime\ 2} \right)}{\upsilon_0 + Vy} dx = \left( \frac{1}{\upsilon_0}\right) \frac{\left( 1 + \frac{1}{2} y^{\prime\ 2} \right)}{1 + k y} dx,
$$
with $k = \frac{V}{\upsilon_0}$.

(c) <br>
$$
\frac{\partial f}{\partial y} = -k \left( 1 + \frac{1}{2}y^{\prime 2} \right) \left( 1 + k y \right)^{-2},
$$
and

$$
\frac{d}{dx} \frac{\partial f}{\partial y^\prime} = \frac{d}{dx} \left( \left( 1 + k y \right)^{-1} y^\prime \right) = -k \left( 1 + k y \right)^{-2} y^{\prime\ 2} + \left( 1 + k y \right)^{-1} y^{\prime\prime}.
$$

So
$$
-k \left( 1 + \frac{1}{2}y^{\prime 2} \right) = -k y^{\prime\ 2} + \left( 1 + k y \right) y^{\prime\prime}.
$$

And finally
$$
\left( 1 + ky \right) y^{\prime\prime} -\frac{1}{2} k y^{\prime\ 2} + k = 0
$$

Taylor suggests the solution $y = \lambda x \left( D - x \right) = \lambda \left( D x - x^2 \right) \Rightarrow y^\prime = \lambda (D - 2 x)$, and $y^{\prime \prime} = -2 \lambda $. 

Plugging into the differential equation: 
$$
\left[ 1 + k \left( \lambda x \left( D - x \right) \right) \right] \left( -2 \lambda \right) - \frac{1}{2} k \left[ \lambda \left( D - 2 x \right) \right]^2 + k = 0
$$

This equation must hold for all $x$, so let's consider $x = 0$ to simplify things, giving (after multplying through by $-1$): $\frac{1}{2} k D^2 \lambda^2 + 2 \lambda - k = 0$.

To solve for lambda, we just apply the quadratic equation:
$$
\boxed{ \lambda = \frac{-2 + \sqrt{ 4 + 2 k^2D^2 }}{k D^2} }
$$

To determine how far north the plane travels, we need to figure out $y$ when $0 = y^\prime = \lambda (D - 2 x) = 0 \Rightarrow x = \frac{1}{2}D$ and $y(x = \frac{1}{2}D) = \lambda \left( \frac{1}{2}D \right) \left( D - \frac{1}{2}D \right) = \frac{1}{4} \lambda D^2 = \boxed{ \frac{\sqrt{4 + 2 k^2 D^2} - 2}{k} }$. 

With $D =$ 2,000 miles, $\upsilon_0 = 500$ mph, $V = 0.5$ mph/mi, max $y \approx$ 400 miles. 

If the plane flew directly to the town along $y = 0$, it would take $T = D/\upsilon_0 =$ 4 hours. Instead, it takes $T = \int_{x = 0}^{D} \left( \frac{ 1 + \frac{1}{2}y^{\prime 2} }{ 1 + k y } \right) \frac{dx}{\upsilon_0}$. Plugging in our solution for $y$ and $y^\prime$, we can do the integral:

```python
import numpy as np
from scipy import integrate as it
import pylab as pl

def func(x, D, v0, k):
    lamb = (np.sqrt(4. + 2.*k**2*D**2) - 2.)/(k*D**2)
    
    y = lamb*x*(D - x)
    yprime = lamb*(D - 2.*x)
    
    return (1. + 0.5*yprime**2)/(1. + k*y)/v0

D = 2000.
v0 = 500.
V = 0.5
k = V/v0
print("The trip takes %3.2f hours, instead of 4 hours." % (it.quad(func, 0., 2000., args = (D, v0, k))[0]))
```

#### 6.25

As usual, we have $dt = \dfrac{ds}{v} = \dfrac{\sqrt{x^{\prime 2} + 1} dy}{v(y)}$ with $x^{\prime} = \dfrac{dx}{dy}$. For the particle falling from rest along the curve from the point $P_0$, the velocity is given by equating the kinetic to the potential energy: 
$$
\frac{1}{2}m v^2 = m g \left( y - y_0 \right),
$$
where we have to incorporate the $y_0$ since the particle will start from rest. Otherwise, the particle would start off with some non-zero velocity.

Thus, we have $v = \sqrt{2 g \left( y - y_0 \right)}$.

So we are interested in this integral:
$$
T = \int_{y = y_0}^{0} \dfrac{\sqrt{x^{\prime 2} + 1}}{\sqrt{2g\left( y - y_0 \right)}}\ dy.
$$

We'll use the variable substitions suggested in Example 6.2 back in the chapter: $y = a \left( 1 - \cos \theta \right)$, $dy = a \sin \theta\ d\theta$, and $x^\prime = \sqrt{\dfrac{1 - \cos \theta}{1 + \cos \theta}}$, giving
$$
T = \sqrt{\dfrac{1}{g}} \int_{\theta = \theta_0}^{\pi} \sqrt{\dfrac{1}{\left( 1 + \cos \theta\right) a \left( \cos \theta_0 - \cos \theta \right) }} a \sin \theta\ d\theta.
$$

Let's do some re-arranging:
$$
\dfrac{1}{\sqrt{ 1 + \cos \theta }} \sin \theta = \sqrt{\dfrac{\sin^2\theta}{1 + \cos \theta }} = \sqrt{\dfrac{1 - \cos^2\theta}{1 + \cos \theta }} = \sqrt{\dfrac{\left( 1 - \cos\theta \right) \left( 1 + \cos\theta \right)}{1 + \cos \theta }} = \sqrt{1 - \cos\theta}.
$$

So
$$
T = \sqrt{\dfrac{a}{g}} \int_{\theta = \theta_0}^{\pi} \sqrt{ \dfrac{1 - \cos \theta}{\cos \theta_0 - \cos \theta} }\ d\theta.
$$

Now we have to actually do the integral. Taylor suggests substituting $\theta = \pi - 2 \alpha$. This substitution means $d\theta = -2 d\alpha$ and changing the integration limits $\int_{\theta = \theta_0}^{\pi} \rightarrow \int_{\alpha = \pi/2 - \theta_0/2}^{0}$.

So we end up with
$$
T = 2\sqrt{\dfrac{a}{g}} \int_{\alpha = 0}^{\pi/2 - \theta_0/2} \sqrt{\dfrac{1 - \cos \left( \pi - 2 \alpha \right) }{\cos \left(\pi - 2 \alpha_0 \right) - \cos \left( \pi - 2 \alpha \right) }}\ d\alpha,
$$
where the integration limits flip because $d\theta$ and $d\alpha$ have opposite signs.

We can use some trig identities to make some substitutions: $\cos \left( \pi - 2 \alpha \right) = -\cos \left( 2\alpha \right) = -1 + 2 \sin^2\alpha $ to give
$$
T = 2 \sqrt{\dfrac{a}{g}} \int_{\alpha = 0}^{\pi/2 - \theta_0/2} \sqrt{\dfrac{1 - \sin^2\alpha}{\sin^2\alpha_0 - \sin^2\alpha}}\ d\alpha = \dfrac{1}{\sqrt{\sin^2\alpha_0 - \sin^2\alpha}}\ \cos \alpha\ d\alpha.
$$

Then Taylor says to try the substitution $u = \sin \alpha$, giving $du = \cos \alpha\ d\alpha$ and
$$
T = 2 \sqrt{\dfrac{a}{g}} \int_{u = 0}^{u_0} \left( \dfrac{du}{\sqrt{u_0^2 - u^2}} \right) = 2 \sqrt{\dfrac{a}{g}} \int_{u = 0}^{u_0} \dfrac{du}{u_0} \dfrac{1}{\sqrt{1 - \left( \frac{u}{u_0} \right)^2}}.
$$

We can integrate with one last substitution: $\dfrac{u}{u_0} = \sin \beta$, which implies $\dfrac{du}{u_0} = \cos \beta\ d\beta$ and
$$
T = 2 \sqrt{\dfrac{a}{g}}  \int_{\beta = 0}^{\pi/2} \dfrac{\cos \beta\ d\beta}{\sqrt{1 - \sin^2 \beta}} = 2 \sqrt{\dfrac{a}{g}}  \int_{\beta = 0}^{\pi/2} d\beta = \boxed{ \pi \sqrt{\dfrac{a}{g}} }.
$$
